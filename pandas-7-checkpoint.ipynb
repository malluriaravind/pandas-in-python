```python
people={
    'first':['aravind','rahul','jake','adam'],
    'last':['reddy','reddy','chris','doe'],
    'email':['malluriaravindreddy@gmail.com','rahul16@gmail.com','jakechris@gmail.com','adam@gmail.com']
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
#how to sort columns 
#sort the values by last name by using sort_value method
df.sort_values(by='last')
```

```python
# if we want to sort this in descending order use ascending=False
df.sort_values(by='last',ascending=False)
```

```python
# if we want to sort on multiple columns
df.sort_values(by=['last','first'])
```

```python
# if we want to sort lastname in desceding order and first name in ascending order
df.sort_values(by=['last','first'],ascending=[False,True],inplace=True)
```

```python
df
```

```python
#if we want the previous dataframe or sort the indexes by ascending order
df.sort_index()
```

```python
# series objects also have the sort() method
# if we want to see simply the last name not entire dataframe
df['last'].sort_values()
```
