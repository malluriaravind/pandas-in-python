```python
#Filtering - Using conditionals to filter rows and columns
people={
    'first':['aravind','rahul','jake'],
    'last':['reddy','reddy','chris'],
    'email':['malluriaravindreddy@gmail.com','rahulreddy@gmail.com','jakechris@gmail.com']
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
df['last']=='reddy'
```

```python
# if we want to access the dataframe with last name contains reddy 
filt=(df['last']=='reddy')
```

```python
df[filt]
#here we had get our dataframe back and returned all of the last name 'reddy'
```

```python
#alternative way
df[df['last']=='reddy']
```

```python
df.loc[filt]
```

```python
# we can use loc and even can grab specific column using .loc
df.loc[filt,'email']
# 1st value filt is rows we want
# 2nd value email is column we want
```

```python
# & -> and , | -> or
filt=(df['last']=='reddy') & (df['first']=='rahul')
```

```python
df.loc[filt,'email']
```

```python
filt=(df['last']=='reddy') | (df['first']=='rahul')
```

```python
df.loc[filt,'email']
```

```python
filt=(df['last']=='reddy')  | (df['first']=='rahul')
```

```python
df.loc[~filt,'email']
```
