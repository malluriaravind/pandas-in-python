```python
#normally we store the details we can use the dictionaries
#represents data for single person
person={
    'first':'aravind',
    'last':'reddy',
    'email':'malluriaravindreddy@gmail.com'
}
```

```python
# reprsents data for multiple persons we will use list inside the dictionary.
people={
    'first':['aravind','rahul','jake'],
    'last':['reddy','panday','chris'],
    'email':['malluriaravindreddy@gmail.com','rahul16@gmail.com','chrisjake@gmail.com']
}
#here we can consider keys as columns and values as rows of the people dictionary.
```

```python
#if we want see the email of all the peoples normally syntax is d['key']
people['email']
```

```python
#dataframes are very similar to this and have more functionality.
import pandas as pd
```

```python
data=pd.DataFrame(people) #the data will be assigned into rows and columns
```

```python
data
```

```python
#accessing the column values of the data frame
data['email']
```

```python
type(data['email'])

#Note:
#data frame is basically rows and columns  (two-dimensional array)
#series is only the rows of particular column (One-dimensional array)
#data frame is basically a container for the multiple series of these objects.
```

```python
#another way to access the column values of data frame
data.email

#note: I always prefer first one for accessing because for example if there is a method called count and if you had a column
# named count and you were trying to access that count column using dot notation that's actually going to access the count
# method from the dataframe instead of that count column
```

```python
# data frames have lot more functionality than standard python.
```

```python
# If we wanted to access multiple columns, now inorder to access multiple columns we can use the bracket notation and pass
# in a list of columns that we want. 
#So If i wanted the lastname and email columns then we would say
data[['last','email']]

#now that we are getting multiple columns this can no longer be a series because series is basically a single column of rows
# so when we get multiple columns like this it's just returning another dataframe and this case it's a filtered down data
# frame with just these specific columns. so we filtered out the first name column and we have last name and email.
```

```python
# if we want to see the columns of the dataframe
data.columns
```

```python
# if we want to get the rows we can use loc and iloc indexes 
#iloc allow us to access rows by integer location, hence the name iloc is integer location.

#if we want to get the first row
data.iloc[0] #returns a series that contains a values of first row of data 

#here index is the column name what those values are, here actually accessing the row it's going to set that index to 
# column name.
```

```python
# we can select multiple columns we can select multiple rows as well by passing in a list of integers if i want 1st and 2nd
# row 
data.iloc[[0,1]]

#by this we can get the two rows of data 
```

```python
# if we want to access the specified rows and columns for eg we want the email addresses of first two rows.
data.iloc[[0,1],2]
```

```python
# with loc we can be searching labels, rows are also called as indexes now we don't have the custom indexes this index is 
# default range of integers.
data
#here 0 1 and 2 these are called labels if i want the first row then by default this just has a label of 0
```

```python
data.loc[0]
```

```python
#with iloc we can use integer to grab  the specific columns but using loc we can use lables for example if we want email
# column of these first two rows no we can just pass the value email.
data.loc[[0,1],'email']
```

```python
# if we want last name and email for first two rows by using loc
data.loc[[0,1],['last','email']]
```

```python
# if we want last name and email for first two rows by using iloc
data.iloc[[0,1],[1,2]]
```
