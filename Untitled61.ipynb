```python
# concatenating all the csv files and making a new csv file 
import os
df=pd.read_csv('Desktop/Sales_Data/Sales_April_2019.csv')
files=[file for file in os.listdir('Desktop/Sales_Data')]
all_data=pd.DataFrame()
for file in files:
    df=pd.read_csv('Desktop/Sales_Data/'+file)
    all_data=pd.concat([all_data,df])
all_data.head()
all_data.to_csv('Desktop/all_month.csv')
```

```python
all_data.head()
```

```python
# renaming the column values
all_data.columns=all_data.columns.str.replace(' ','_')
```

```python
all_data
```

```python
# check for dataframe contains nullvalue or not
all_data.isnull().any()
```

```python
# we have null values in the dataframe if we want to remove the missing values
all_data=all_data.dropna(axis=0,how='all')
```

```python
## check for dataframe contains nullvalue or not
all_data.isnull().any()
```

```python
# change the datatype of columns
all_data['Quantity_Ordered']=pd.to_numeric(all_data['Quantity_Ordered'],errors='coerce')
all_data['Price_Each']=pd.to_numeric(all_data['Price_Each'],errors='coerce')
```

```python
all_data.dtypes
```

```python
# to get the total sum of each product sold by customers
all_data.groupby('Product').sum()['Price_Each']
```

```python
# to get the count of each product sold by customers
all_data.groupby('Product').count()['Quantity_Ordered']
```

```python
#creating a new column using slicing
all_data=all_data[all_data['Order_Date'].str[0:2]!='Or']
all_data['Month']=all_data['Order_Date'].str[0:2]
all_data['Month']=pd.to_numeric(all_data['Month'])
all_data.head()
```

```python
# total cost spend by customers for each order
all_data['Sales']=all_data['Quantity_Ordered']*all_data['Price_Each']
```

```python
all_data
```

```python
# apply() Pandas. apply allow the users to pass a function and apply it on every single value of the Pandas series.
all_data['State']=all_data['Purchase_Address'].apply(lambda x:x.split(',')[2][0:3])
```

```python
all_data.head()
```

```python
# apply() Pandas. apply allow the users to pass a function and apply it on every single value of the Pandas series.
all_data['Zipcode']=all_data['Purchase_Address'].apply(lambda x:x.split(',')[2][3:])
```

```python
all_data.head()
```

```python
# apply() Pandas. apply allow the users to pass a function and apply it on every single value of the Pandas series.
all_data['City']=all_data['Purchase_Address'].apply(lambda x:x.split(',')[1])
```

```python
all_data.head()
```

```python
# get the five state which has maximum sales
all_data.groupby('State').sum()['Sales'].sort_values(ascending=False).head()
```

```python
# catgorical plot - bar plot for visualising sales for each state
state_group=all_data.groupby('State')
sales=state_group.sum()['Sales']
states=[state for state,df in state_group]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
sns.barplot(states,sales)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# which month has maximum no of sales
all_data.groupby('Month').sum()['Sales']
```

```python
# visualize which month has maximum no of sales
months=all_data.groupby('Month')
sales=months.count()['Sales']
all_months=[month for month,df in months]
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
sns.lineplot(all_months,sales,marker='o')
plt.xticks(all_months)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid()
plt.show()
```
