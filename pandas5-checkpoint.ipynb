```python
people={
    'first':['aravind','rahul','jake'],
    'last':['reddy','reddy','chris'],
    'email':['malluriaravindreddy@gmail.com','rahul16@gmail.com','jakechris@gmail.com']
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
#updating the column names
df.columns=['first name','last name','email address']
```

```python
df
```

```python
#use the list comprehension to convert the all column names in to upper
df.columns=[x.upper() for x in df.columns]
```

```python
df
```

```python
#replacing the spaces with underscore
df.columns=df.columns.str.replace(' ','_')
```

```python
#convert the all uppercase column names to lower case
df.columns=[x.lower() for x in df.columns]
```
